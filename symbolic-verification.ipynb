# Symbolic Verifications to Proofs

## 1. Validity of metric spaces


We symbolically show that tight examples in Theorem 5.7 to Theorem 5.9 consist a valid metric space. We show that the 3-path triangle inequalities hold symbolically.

### Theorem 5.7

```python
import sympy as sp
import pandas as pd

def reduce_alpha(expr, poly):
    """
    Reduce a polynomial expression modulo
    alpha^3 - alpha^2 - alpha - 1 = 0.
    """
    return sp.Poly(sp.expand(expr), alpha, domain="EX").rem(
        sp.Poly(poly, alpha, domain="EX")
    ).as_expr()

# Symbol
alpha = sp.Symbol("alpha_4", real=True)

# Relation: alpha_4^3 - alpha_4^2 - alpha_4 - 1 = 0
P4 = alpha**3 - alpha**2 - alpha - 1

# gamma = alpha_4^3 - 2 alpha_4
gamma = alpha**3 - 2*alpha
```

```python
# Rows are i_1, i_2, i_3.
# Columns are c_0, c_1, c_2, c_3.

D = {
    (1, 0): alpha**3,
    (1, 1): alpha**2,
    (1, 2): alpha,
    (1, 3): 1,

    (2, 0): alpha,
    (2, 1): 1,
    (2, 2): alpha**3,
    (2, 3): alpha**2,

    (3, 0): alpha,
    (3, 1): alpha*gamma,
    (3, 2): gamma,
    (3, 3): alpha**2,
}

# Reduce every entry using alpha^3 = alpha^2 + alpha + 1
D = {key: reduce_alpha(value, P4) for key, value in D.items()}
```

```python
rows = [1, 2, 3]
cols = [0, 1, 2, 3]

residuals = []

for a in rows:
    for ap in rows:
        for b in cols:
            for bp in cols:
                residual = reduce_alpha(
                    D[a, b] + D[ap, b] + D[ap, bp] - D[a, bp], P4
                )

                residuals.append({
                    "a": a,
                    "ap": ap,
                    "b": b,
                    "bp": bp,
                    "residual": sp.factor(residual),
                })

df = pd.DataFrame(residuals)

# The positive real root of alpha^3 - alpha^2 - alpha - 1
alpha_exact = sp.CRootOf(P4, 0)

def is_nonnegative_mod_q(expr):
    expr = reduce_alpha(expr, P4)
    value = expr.subs(alpha, alpha_exact)
    return value.is_nonnegative

df["holds?"] = df["residual"].apply(is_nonnegative_mod_q)

violations = df[df["holds?"] != True]
print(f"Number of violations: {len(violations)}")
```

As we see that the number of violations for the 3-path-inequalities are zero, the example consists a metric space.

### Theorem 5.8

```python
# Symbol
alpha = sp.Symbol("alpha_5", real=True)

# Relation: alpha_5^5 - 3 alpha_5^4 + 3 alpha_5^3 - 3 alpha_5^2 + 3 alpha_5 - 3 = 0
P5 = alpha**5 - 3*alpha**4 + 3*alpha**3 - 3*alpha**2 + 3*alpha - 3

# delta = alpha_5 - 1
delta = alpha - 1
```

```python
# Rows are i_1,...,i_5.
# Columns are c_0,...,c_4.

D = {
    (1, 0): alpha**2 * delta**4,
    (1, 1): alpha * delta**4,
    (1, 2): delta**4,
    (1, 3): 2 * alpha * delta**3,
    (1, 4): 2 * delta**3,

    (2, 0): 2 * delta**2,
    (2, 1): alpha**2 * delta**3,
    (2, 2): alpha * delta**3,
    (2, 3): delta**3,
    (2, 4): 2 * alpha * delta**2,

    (3, 0): 2 * alpha * delta,
    (3, 1): 2 * delta,
    (3, 2): alpha**2 * delta**2,
    (3, 3): alpha * delta**2,
    (3, 4): delta**2,

    (4, 0): delta,
    (4, 1): 2 * alpha,
    (4, 2): alpha**2 * delta**4 - delta**4 - delta,
    (4, 3): alpha**2 * delta,
    (4, 4): alpha * delta,

    (5, 0): alpha,
    (5, 1): 1,
    (5, 2): alpha**2 * delta**3 - delta**3 + 1,
    (5, 3): alpha**2 * delta**3 - delta**3 - 1,
    (5, 4): alpha**2,
}

# Reduce all entries modulo P5
D = {key: sp.factor(reduce_alpha(value, P5)) for key, value in D.items()}
```

```python
rows = [1, 2, 3, 4, 5]
cols = [0, 1, 2, 3, 4]

table = pd.DataFrame(
    [[D[r, c] for c in cols] for r in rows],
    index=[f"i_{r}" for r in rows],
    columns=[f"c_{c}" for c in cols],
)

residuals = []

for a in rows:
    for ap in rows:
        for b in cols:
            for bp in cols:
                residual = reduce_alpha(
                    D[a, b] + D[ap, b] + D[ap, bp] - D[a, bp], P5
                )

                residuals.append({
                    "a": a,
                    "ap": ap,
                    "b": b,
                    "bp": bp,
                    "residual": sp.factor(residual),
                })

df = pd.DataFrame(residuals)

# The unique real root of P5
alpha_exact = sp.CRootOf(P5, 0)

def is_nonnegative_mod_P5(expr):
    expr = reduce_alpha(expr, P5)
    value = expr.subs(alpha, alpha_exact)
    return value.is_nonnegative


df["holds?"] = df["residual"].apply(is_nonnegative_mod_P5)

violations = df[df["holds?"] != True]
print(f"Number of violations: {len(violations)}")
```

As we see that the number of violations for the 3-path-inequalities are zero, the example consists a metric space.

### Theorem 5.9

```python
alpha = sp.Symbol("alpha_6", real=True)

# Defining polynomial
P6 = alpha**5 - 2*alpha**4 + alpha**2 - 4*alpha + 2

gamma = alpha - 1
delta = alpha**2 - alpha + 2
xi = alpha**3 * (alpha - 1)**2 - alpha**2 * (alpha - 1) - alpha
```

```python
D = {
    (0, 0): gamma * delta,
    (0, 1): alpha**3 * gamma**2,
    (0, 2): alpha**2 * gamma**2,
    (0, 3): alpha * gamma**2,
    (0, 4): alpha**2 * gamma,
    (0, 5): alpha * gamma * delta,

    (1, 0): alpha**2,
    (1, 1): alpha,
    (1, 2): alpha**3,
    (1, 3): alpha * xi,
    (1, 4): xi,
    (1, 5): alpha**3,

    (2, 0): alpha * xi,
    (2, 1): xi,
    (2, 2): alpha**3,
    (2, 3): alpha**2,
    (2, 4): alpha,
    (2, 5): alpha**3 - 2 * alpha * gamma,

    (3, 0): alpha * gamma**2,
    (3, 1): alpha**2 * gamma,
    (3, 2): alpha * gamma * delta,
    (3, 3): gamma * delta,
    (3, 4): alpha**3 * gamma**2,
    (3, 5): alpha**2 * gamma**2,

    (4, 0): alpha**3 * gamma,
    (4, 1): alpha**2 * gamma,
    (4, 2): alpha * gamma,
    (4, 3): alpha**3 * gamma,
    (4, 4): alpha**2 * gamma,
    (4, 5): alpha * gamma,
}

# Reduce every table entry modulo P6
D = {key: sp.factor(reduce_alpha(value, P6)) for key, value in D.items()}
```

```python
rows = [0, 1, 2, 3, 4]
cols = [0, 1, 2, 3, 4, 5]

table = pd.DataFrame(
    [[D[r, c] for c in cols] for r in rows],
    index=[f"i_{r}" for r in rows],
    columns=[f"c_{c}" for c in cols],
)

residuals = []

for a in rows:
    for ap in rows:
        for b in cols:
            for bp in cols:
                residual = reduce_alpha(
                    D[a, b] + D[ap, b] + D[ap, bp] - D[a, bp], P6
                )

                residuals.append({
                    "a": a,
                    "ap": ap,
                    "b": b,
                    "bp": bp,
                    "residual": sp.factor(residual),
                })

df = pd.DataFrame(residuals)

# Root in [2, infinity)
alpha_exact = sp.CRootOf(P6, 2)

def is_nonnegative_mod_P6(expr):
    expr = reduce_alpha(expr, P6)
    value = expr.subs(alpha, alpha_exact)
    return value.is_nonnegative


df["holds?"] = df["residual"].apply(is_nonnegative_mod_P6)

violations = df[df["holds?"] != True]
print(f"Number of violations: {len(violations)}")
```

As we see that the number of violations for the 3-path-inequalities are zero, the example consists a metric space.

## 2. Validity of multipliers

We symbolically show that the calculations of Theorem 5.8 and Lemmas C.1 to C.6 are correct.

### Theorem 5.8

```python
import sympy as sp

# Formal labels
i = sp.Symbol("i")
ip = sp.Symbol("i'")
i2 = sp.Symbol("i_2")

c = sp.symbols("c_0:5")

# Formal distance function
d = sp.Function("d")

def D(a, b):
    return d(a, c[b])


Phi = (
    1*(D(i,2) + D(i2,2) + D(i2,3) - D(i,3))
    + 8*(D(i,2) + D(ip,2) + D(ip,4) - D(i,4))
    + 12*(D(ip,2) + D(i,2) + D(i,1) - D(ip,1))
    + 1*(D(ip,2) + D(i,2) + D(i,3) - D(ip,3))
    + 3*(D(ip,2) + D(i2,2) + D(i2,3) - D(ip,3))
    + 6*(D(i2,2) + D(i,2) + D(i,1) - D(i2,1))
    + 8*(D(i,4) - 2*D(i,0))
    + 16*(D(i,0) - 2*D(i,1))
    + 14*(D(i,1) - 2*D(i,2))
    + 12*(D(ip,1) - 2*D(ip,2))
    + 4*(D(ip,3) - 2*D(ip,4))
    + 6*(D(i2,1) - 2*D(i2,2))
    + 2*(D(i2,2) - 2*D(i2,3))
)

Phi_reduced = sp.expand(Phi)
print(Phi_reduced == 0)
```

Hence, the first part using multipliers is correct. Next, we check the latter part of Theorem 5.8.

```python
import sympy as sp

alpha = sp.Symbol("alpha", real=True)

# labels i_2, i_3, i_4, i and c_0,...,c_5
i2, i3, i4, i = sp.symbols("i_2 i_3 i_4 i")
c = sp.symbols("c_0:6")

# formal distance function
dist = sp.Function("d")

def D(a, b):
    return dist(a, c[b])


# polynomial p(alpha)
p = (
    alpha**8
    - 4*alpha**7
    + 7*alpha**6
    - 8*alpha**5
    + 6*alpha**4
    - 4*alpha**3
    + alpha**2
    - 2*alpha
    - 1
)


# 3-path residuals
T1 = D(i2, 2) + D(i,  2) + D(i,  1) - D(i2, 1)
T2 = D(i3, 4) + D(i,  4) + D(i,  2) - D(i3, 2)
T3 = D(i4, 4) + D(i3, 4) + D(i3, 3) - D(i4, 3)
T4 = D(i,  2) + D(i2, 2) + D(i2, 3) - D(i,  3)
T5 = D(i,  4) + D(i4, 4) + D(i4, 5) - D(i,  5)


# deviation residuals
Delta1  = D(i2, 1) - alpha*D(i2, 2)
Delta2  = D(i2, 2) - alpha*D(i2, 3)
Delta3  = D(i3, 2) - alpha*D(i3, 3)
Delta4  = D(i3, 3) - alpha*D(i3, 4)
Delta5  = D(i4, 3) - alpha*D(i4, 4)
Delta6  = D(i4, 4) - alpha*D(i4, 5)
Delta7  = D(i,  5) - alpha*D(i,  0)
Delta8  = D(i,  0) - alpha*D(i,  1)
Delta9  = D(i,  1) - alpha*D(i,  2)
Delta10 = D(i,  3) - alpha*D(i,  4)


# K and L
K = alpha**3 - 2*alpha**2 + 2*alpha + 1
L = alpha**5*(alpha - 1)**3 - (alpha + 1)*K


# coefficients
A = (alpha + 1)*K
B = alpha**2*(alpha - 1)*(alpha + 1)
C = alpha**2*(alpha - 1)**2*(alpha + 1)
Dcoef = alpha*(alpha - 1)*K
Ecoef = alpha**3*(alpha - 1)**3
F = (alpha - 1)*K
G = alpha**2*(alpha - 1)**3
H = alpha**4*(alpha - 1)**3


# weighted sum Phi
Phi = (
    A*(T1 + Delta1)
    + B*(T2 + Delta3 + Delta4)
    + C*(T3 + Delta5)
    + Dcoef*(T4 + Delta10)
    + Ecoef*(T5 + Delta7)
    + F*Delta2
    + G*Delta6
    + H*Delta8
    + L*Delta9
)


# reduce modulo p(alpha)
def reduce_alpha(expr):
    return sp.Poly(sp.expand(expr), alpha, domain="EX").rem(
        sp.Poly(p, alpha, domain="EX")
    ).as_expr()


Phi_reduced = sp.factor(reduce_alpha(Phi))

print("Phi reduces modulo p(alpha) to 0:", Phi_reduced == 0)
```

The following checks that the coefficients are strictly positive.

```python
import sympy as sp

alpha = sp.Symbol("alpha", real=True)

p = (
    alpha**8
    - 4*alpha**7
    + 7*alpha**6
    - 8*alpha**5
    + 6*alpha**4
    - 4*alpha**3
    + alpha**2
    - 2*alpha
    - 1
)

P = sp.Poly(p, alpha, domain=sp.QQ)

# exact rational interval containing alpha_bar
a = sp.Rational(201, 100)
b = sp.Rational(202, 100)
mid = (a + b) / 2

K = alpha**3 - 2*alpha**2 + 2*alpha + 1
L = alpha**5*(alpha - 1)**3 - (alpha + 1)*K

coefficients = [
    (alpha + 1)*K,
    alpha**2*(alpha - 1)*(alpha + 1),
    alpha**2*(alpha - 1)**2*(alpha + 1),
    alpha*(alpha - 1)*K,
    alpha**3*(alpha - 1)**3,
    (alpha - 1)*K,
    alpha**2*(alpha - 1)**3,
    alpha**4*(alpha - 1)**3,
    L,
]

# exact symbolic checks
root_isolated = P.count_roots(a, b) == 1

all_positive = root_isolated and all(
    sp.Poly(sp.expand(q), alpha, domain=sp.QQ).count_roots(a, b) == 0
    and sp.sign(sp.Poly(sp.expand(q), alpha, domain=sp.QQ).eval(mid)) == 1
    for q in coefficients
)

print("All coefficients positive at alpha_bar:", all_positive)
```

```python
import sympy as sp

alpha = sp.Symbol("alpha", real=True)

# labels i, i', i_2, i_5 and c_0,...,c_5
i, ip, i2, i5 = sp.symbols("i i_prime i_2 i_5")
c = sp.symbols("c_0:6")

# formal distance function
dist = sp.Function("d")

def D(a, b):
    return dist(a, c[b])


# 3-path residuals
T1 = D(i,  4) + D(ip, 4) + D(ip, 2) - D(i,  2)
T2 = D(i,  1) + D(ip, 1) + D(ip, 5) - D(i,  5)
T3 = D(ip, 5) + D(i5, 5) + D(i5, 0) - D(ip, 0)
T4 = D(ip, 2) + D(i2, 2) + D(i2, 3) - D(ip, 3)
T5 = D(i2, 3) + D(i,  3) + D(i,  1) - D(i2, 1)
T6 = D(i5, 0) + D(i,  0) + D(i,  4) - D(i5, 4)


# deviation residuals
Delta1  = D(i,  5) - alpha*D(i,  0)
Delta2  = D(i,  0) - alpha*D(i,  1)
Delta3  = D(i,  2) - alpha*D(i,  3)
Delta4  = D(i,  3) - alpha*D(i,  4)

Delta5  = D(ip, 0) - alpha*D(ip, 1)
Delta6  = D(ip, 1) - alpha*D(ip, 2)
Delta7  = D(ip, 3) - alpha*D(ip, 4)
Delta8  = D(ip, 4) - alpha*D(ip, 5)

Delta9  = D(i2, 1) - alpha*D(i2, 2)
Delta10 = D(i2, 2) - alpha*D(i2, 3)

Delta11 = D(i5, 4) - alpha*D(i5, 5)
Delta12 = D(i5, 5) - alpha*D(i5, 0)


# Phi
Phi = (
    T1 + T2 + T3 + T4 + T5 + T6
    + Delta1 + Delta2 + Delta3 + Delta4
    + Delta5 + Delta6 + Delta7 + Delta8
    + Delta9 + Delta10 + Delta11 + Delta12
)


# claimed expansion
claimed = (2 - alpha) * (
    D(i,  0) + D(i,  1) + D(i,  3) + D(i,  4)
    + D(ip, 1) + D(ip, 2) + D(ip, 4) + D(ip, 5)
    + D(i2, 2) + D(i2, 3)
    + D(i5, 0) + D(i5, 5)
)


print("Claimed expansion holds:", sp.expand(Phi - claimed) == 0)
```

### Lemma C.3

```python
import sympy as sp

alpha = sp.Symbol("alpha", real=True)

# labels i_1, i_4, i', i and c_0,...,c_5
i1, i4, ip, i = sp.symbols("i_1 i_4 i_prime i")
c = sp.symbols("c_0:6")

# formal distance function
dist = sp.Function("d")

def D(a, b):
    return dist(a, c[b])


# polynomial p(alpha)
p = (
    alpha**8
    - 4*alpha**7
    + 7*alpha**6
    - 8*alpha**5
    + 6*alpha**4
    - 6*alpha**3
    + 3*alpha**2
    - 2*alpha
    + 1
)


# 3-path residuals
T1 = D(i1, 2) + D(ip, 2) + D(ip, 0) - D(i1, 0)
T2 = D(i4, 4) + D(i,  4) + D(i,  3) - D(i4, 3)
T3 = D(ip, 0) + D(i,  0) + D(i,  1) - D(ip, 1)
T4 = D(ip, 0) + D(i,  0) + D(i,  4) - D(ip, 4)
T5 = D(i,  1) + D(i1, 1) + D(i1, 2) - D(i,  2)
T6 = D(i,  4) + D(i4, 4) + D(i4, 5) - D(i,  5)


# deviation residuals
Delta1  = D(i1, 0) - alpha*D(i1, 1)
Delta2  = D(i1, 1) - alpha*D(i1, 2)
Delta3  = D(i4, 3) - alpha*D(i4, 4)
Delta4  = D(ip, 1) - alpha*D(ip, 2)
Delta5  = D(ip, 4) - alpha*D(ip, 5)
Delta6  = D(i,  2) - alpha*D(i,  3)
Delta7  = D(i,  5) - alpha*D(i,  0)
Delta8  = D(i4, 4) - alpha*D(i4, 5)
Delta9  = D(ip, 5) - alpha*D(ip, 0)
Delta10 = D(i,  0) - alpha*D(i,  1)
Delta11 = D(i,  3) - alpha*D(i,  4)


# P and Q
P = alpha**3 - alpha**2 + 2*alpha - 1
Q = (
    alpha**5
    - 3*alpha**4
    + 4*alpha**3
    - 3*alpha**2
    + 3*alpha
    - 1
)


# coefficients
A = alpha**4*(alpha - 1)**2
B = (alpha + 1)*P
C = alpha**3*(alpha - 1)**2
Dcoef = alpha**3*(alpha - 1)
E = alpha**4*(alpha - 1)**3
F = alpha*(alpha - 1)*P
G = (alpha - 1)*P
H = alpha**4*(alpha - 1)
J = alpha**2*(alpha - 1)**2*(alpha**2 - alpha + 1)
K = alpha*(alpha - 1)*Q


# weighted sum Phi
Phi = (
    A*(T1 + Delta1 + Delta2)
    + B*(T2 + Delta3)
    + C*(T3 + Delta4)
    + Dcoef*(T4 + Delta5)
    + E*(T5 + Delta6)
    + F*(T6 + Delta7)
    + G*Delta8
    + H*Delta9
    + J*Delta10
    + K*Delta11
)


# reduce modulo p(alpha)
def reduce_alpha(expr):
    return sp.Poly(sp.expand(expr), alpha, domain="EX").rem(
        sp.Poly(p, alpha, domain="EX")
    ).as_expr()


Phi_reduced = sp.factor(reduce_alpha(Phi))

print("Phi reduces modulo p(alpha) to 0:", Phi_reduced == 0)
```
